# Libraries that we will need

```python
from yahoo_fin.stock_info import get_data
import pandas as pd
import yahoo_fin.stock_info as si
import yahoo_fin.options as ops
```

# Downloading One Ticket

```python
Historical_Data = get_data("amzn", start_date="12/04/2009", end_date="12/04/2019", index_as_date = True, interval="1d")
Historical_Data
```

# Downloading Several Tickets

```python
tickers = ["goog", "ko", "mmm"]

Historical_Data = {}
for t in tickers:
    Historical_Data[t] = get_data(t)
    
Historical_Data
```

# Getting the tickets that trade inside an especific index

```python
dow_companies = si.tickers_dow()

print("Companies in the Dow Jones:", len(dow_companies))
dow_companies[0:15]
```

```python
sp_companies = si.tickers_sp500()

print("Companies in sp500:", len(sp_companies))
sp_companies[0:15]
```

```python
nasdaq_companies = si.tickers_nasdaq()

print("Companies in the NASDAQ:", len(nasdaq_companies))
nasdaq_companies[0:15]
```
